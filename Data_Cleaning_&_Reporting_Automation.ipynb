```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Salary_Dataset_DataScienceLovers.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna("Unknown", inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
print(df.describe())
```

```python
salary_by_role = df.groupby('Job Roles')['Salary'].mean()

salary_by_role.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Average Salary by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Salary")

plt.show()
```

```python
df.to_csv(
    "Cleaned_Salary_Data.csv",
    index=False
)
```
